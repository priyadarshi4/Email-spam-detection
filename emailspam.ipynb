```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/email.csv')
df.head(5)
```

```python
df["Category"] = df["Category"].map({"ham":0,"spam":1})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.pipeline import make_pipeline , FeatureUnion
from sklearn.linear_model import LogisticRegression
from sklearn.compose import make_column_transformer
```

```python
df = df.dropna(subset=['Category'])
```

```python
x = df[["Message"]]
y = df["Category"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
c1 = make_column_transformer((TfidfVectorizer(),'Message'))
```

```python
c2 = LogisticRegression()
```

```python
from re import M
pipelines = make_pipeline(c1,c2)
pipelines2 = make_pipeline(c1,MultinomialNB())
```

```python
pipelines.fit(x_train,y_train)
pipelines2.fit(x_train,y_train)
```

```python
y_pred = pipelines.predict(x_test)
y_pred2 = pipelines2.predict(x_test)
```

```python
result = accuracy_score(y_test,y_pred)
result2 = accuracy_score(y_test,y_pred2)
```
